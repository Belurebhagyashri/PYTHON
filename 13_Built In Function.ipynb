# Built-in Functions with Examples

```python
min, max, sum, sorted, reversed
```

# 1.len(list)

```python
Returns the number of items in the list.
```

```python
fruits = ["apple", "banana", "cherry"]
length = len(fruits)
print(length) 
```

# 2.sorted(iterable)

```python
>> Returns a new sorted list from the items in the iterable (doesn't modify the original list).
```

```python
numbers = [3, 1, 4, 1, 5, 9, 2, 6]
sorted_numbers = sorted(numbers)
print(numbers)       
print(sorted_numbers) 
```

```python
numbers = [3, 1, 4, 1, 5, 9, 2, 6]
print("length of list is", len(numbers))
print("sum of item is list is", sum(numbers))
print("lowest item in the is is", min(numbers))
print("highest item in the list is",max(numbers))
```

# 3.sum(list)

```python
Returns the sum of all numeric items in the list.
```

```python
numbers = [1, 2, 3, 4, 5]
total = sum(numbers)
print(total)  
```

# 4.min(list)

```python
Returns the smallest item in the list.
```

```python
numbers = [3, 1, 4, 1, 5, 9, 2, 6]
smallest = min(numbers)
print(smallest)  
```

# 5.max(list)

```python
Returns the largest item in the list.
```

```python
numbers = [3, 1, 4, 1, 5, 9, 2, 6]
largest = max(numbers)
print(largest)  
```
